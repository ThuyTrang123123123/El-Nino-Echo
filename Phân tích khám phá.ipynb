## **1. Giới thiệu dự án**

### **1.1. Bối cảnh nghiên cứu**
Biến đổi khí hậu và hiện tượng nóng lên toàn cầu không chỉ dừng lại ở các chỉ số nhiệt độ bề mặt đại dương (SST) khô khan, mà đang tác động trực tiếp đến an ninh lương thực và chuỗi cung ứng toàn cầu. Trong đó, El Niño – pha nóng của  chu kỳ ENSO (El Niño-Southern Oscillation) – là một trong những hình thái khí hậu cực đoan mang tính hủy diệt cao nhất, gây ra hạn hán nghiêm trọng tại các vựa nông sản lớn trên thế giới như Đông Nam Á và Nam Mỹ. Sự đứt gãy nguồn cung này tất yếu dẫn đến những "cú sốc" về giá cả hàng hóa, đặc biệt là các mặt hàng thiết yếu như Gạo và Cà phê.


### **1.2. Tuyên bố vấn đề**
Trong kinh tế học nông nghiệp và khí hậu, tác động của nhiệt độ lên giá cả không diễn ra ngay lập tức. Có một khoảng thời gian "độ trễ" (lag time) kể từ khi chỉ số El Niño (ONI) đạt đỉnh cho đến khi nguồn cung thực sự cạn kiệt và giá nông sản lập đỉnh mới trên thị trường. Việc không xác định định lượng được độ trễ này khiến các nhà hoạch định chính sách và doanh nghiệp gặp khó khăn trong việc dự phòng rủi ro.
Dự án này được thực hiện nhằm trả lời một câu hỏi cốt lõi bằng dữ liệu: "Mất bao nhiêu tháng kể từ khi một chu kỳ El Niño cực đoan bùng phát để giá Gạo và Cà phê toàn cầu thực sự gánh chịu hậu quả, và mức độ tương quan là bao nhiêu?"


### **1.3. Mục tiêu dự án**
Dự án được triển khai với sự cẩn trọng và tỉ mỉ qua từng bước xử lý số liệu, tập trung vào 3 mục tiêu chính:
Thống kê mô tả: Làm sạch và hợp nhất các chuỗi dữ liệu lịch sử về khí hậu (Chỉ số ONI) và kinh tế vĩ mô (Giá Gạo, Giá Cà phê) trong nhiều thập kỷ.
Phân tích độ trễ (Lag Analysis): Ứng dụng kỹ thuật dịch chuyển chuỗi thời gian để đo lường hệ số tương quan Pearson giữa biến đổi khí hậu và biến động giá cả qua từng khung thời gian (3 tháng, 6 tháng, 9 tháng, 12 tháng).
Kể chuyện bằng dữ liệu (Data Storytelling): Xây dựng hệ thống biểu đồ trực quan hóa đa trục nhằm phác họa rõ nét mối liên hệ nhân quả và độ trễ pha giữa hai tập dữ liệu, từ đó cung cấp góc nhìn thực tiễn cho bài toán dự phòng rủi ro nông nghiệp.

### **2.1. Tiền xử lý & Hợp nhất dữ liệu**

```python
# Xoay trục dữ liệu World Bank
df_wb_wide_raw = df_wb_raw.pivot_table(
    index='date',
    columns='commodity_name',
    values='price_nominal_usd'
).reset_index()

df_wb_wide_raw.rename(columns={
    'date': 'Date',
    'Rice, Thai 5%': 'Rice',
    'Coffee, Robusta': 'Robusta',
    'Coffee, Arabica': 'Arabica'
}, inplace=True)

# Lọc cột để tối ưu hóa bộ nhớ
df_enso = df_enso_raw[['Date', 'ONI']].copy()
df_wb_wide = df_wb_wide_raw[['Date', 'Rice', 'Robusta', 'Arabica']].copy()

# Ép kiểu String trong file csv sang Datetime để đồng bộ
df_enso['Date'] = pd.to_datetime(df_enso['Date'])
df_wb_wide['Date'] = pd.to_datetime(df_wb_wide['Date'])

# Hợp nhất Dữ liệu
df_merged = pd.merge(df_enso, df_wb_wide, on='Date', how='inner')

# Nội suy bù đắp giá trị khuyết thiếu
df_merged['Rice'] = df_merged['Rice'].interpolate(method='linear')
df_merged['Robusta'] = df_merged['Robusta'].interpolate(method='linear')
df_merged['Arabica'] = df_merged['Arabica'].interpolate(method='linear')

# Lọc bỏ các dòng NaN còn sót lại (nếu có)
df_final_data = df_merged.dropna()

#df_final.to_csv('cleaned_data_worldbank.csv', index=False)
```

### **2.2. Phân tích khám phá**

Trước khi đưa dữ liệu vào mô hình thuật toán, nhóm thực hiện **thống kê mô tả** nhằm đánh giá
cấu trúc tổng thể, biên độ dao động và tính ổn định của các chuỗi lịch sử. Mục tiêu là phát
hiện các điểm dị biệt (outliers) hoặc bất thường trong bộ dữ liệu World Bank có thể làm nhiễu
kết quả phân tích độ trễ ở các bước tiếp theo.

**Giả thuyết nền tảng:** Tác động của sốc khí hậu lên giá nông sản **không** diễn ra tức thời.

Chuỗi nhân quả vật lý cần thời gian để truyền dẫn:

> Đại dương nóng lên (El Niño) → Đảo lộn luồng gió toàn cầu → Hạn hán tại Đông Nam Á &
> Nam Mỹ → Mất mùa, giảm sản lượng → Khan hiếm nguồn cung → Giá tăng trên thị trường quốc tế

Mỗi mắt xích trên cần nhiều tháng để xảy ra. Vì vậy, khi đo tương quan tại cùng một thời điểm
(Lag = 0), kết quả kỳ vọng phải **rất gần 0** — gần như không có mối liên hệ tức thời giữa ONI
và giá cả.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = df_final_data.copy()
df.head()
```

```python
# Thống kê mô tả
cols = ['ONI', 'Rice', 'Robusta']

desc = df[cols].describe().round(3)

desc.index = ['Số quan sát', 'Trung bình', 'Độ lệch chuẩn',
              'Min', 'Q1 (25%)', 'Trung vị (50%)', 'Q3 (75%)', 'Max']

desc.columns = ['Chỉ số ONI (El Niño)', 'Giá Gạo (USD/tấn)', 'Giá Cà phê Robusta (USD/tấn)']
desc
```

```python
### Ma trận tương quan & Heatmap (Lag = 0)
```

```python
# Ma trận tương quan Pearson tại Lag = 0
corr_matrix = df[cols].corr(method='pearson').round(3)
corr_matrix.index   = ['Chỉ số ONI', 'Giá Gạo', 'Giá Cà phê Robusta']
corr_matrix.columns = ['Chỉ số ONI', 'Giá Gạo', 'Giá Cà phê Robusta']

print(corr_matrix.to_string())
print()
r_rice    = df['ONI'].corr(df['Rice'])
r_robusta = df['ONI'].corr(df['Robusta'])
print(f"  Tương quan ONI <-> Giá Gạo:    r = {r_rice:.3f}")
print(f"  Tương quan ONI <-> Giá Cà phê: r = {r_robusta:.3f}")

# Vẽ Heatmap
fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(
    corr_matrix,
    annot=True, fmt='.3f',
    cmap='coolwarm', vmin=-1, vmax=1,
    linewidths=0.5, linecolor='white',
    annot_kws={'size': 13, 'weight': 'bold'},
    square=True, ax=ax
)
ax.set_title('Heatmap Tương quan Tức thời (Lag = 0)\nONI vs Giá Gạo & Cà phê Robusta',
             fontsize=13, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('heatmap_zero_lag.png', dpi=150, bbox_inches='tight')
plt.show()
```

Kết quả Heatmap xác nhận hệ số tương quan Pearson giữa ONI và giá Gạo/Cà phê tại Lag = 0
đều nằm trong khoảng **[-0.1 ; 0.1]** — gần như **không có tương quan tức thời**.

Giả thuyết đã được chứng minh bằng dữ liệu: Một chu kỳ El Niño bùng phát không gây ra cú
sốc giá ngay lập tức. Tác động cần một khoảng thời gian trễ (lag time) đủ dài để truyền dẫn
qua toàn bộ chuỗi cung ứng vật lý. Bước tiếp theo sẽ áp dụng kỹ thuật time-shifting để tìm
khung thời gian cho hệ số tương quan cao nhất.

##2.3 — Kỹ thuật Phân tích Độ trễ
Đây là thuật toán cốt lõi của dự án. Nhóm áp dụng kỹ thuật **dịch chuyển chuỗi thời gian
(time-shifting)**: đẩy lùi chuỗi giá Gạo và Cà phê về tương lai theo từng bước 1 tháng,
từ Lag = 1 đến Lag = 12. Tại mỗi bước, hệ số tương quan Pearson giữa ONI và giá (đã dịch)
được tính lại. **Đỉnh của đường cong tương quan** chỉ đích danh số tháng mà El Niño tác
động mạnh nhất đến giá nông sản.

```python
# Quét tương quan từ Lag = 0 đến Lag = 12
lags              = range(0, 13)
corr_rice_lags    = []
corr_robusta_lags = []

for lag in lags:
    shifted_rice     = df['Rice'].shift(-lag)
    shifted_robusta  = df['Robusta'].shift(-lag)
    corr_rice_lags.append(df['ONI'].corr(shifted_rice))
    corr_robusta_lags.append(df['ONI'].corr(shifted_robusta))

# Xác định đỉnh tương quan
best_lag_rice    = list(lags)[np.argmax(np.abs(corr_rice_lags))]
best_lag_robusta = list(lags)[np.argmax(np.abs(corr_robusta_lags))]
best_r_rice      = corr_rice_lags[best_lag_rice]
best_r_robusta   = corr_robusta_lags[best_lag_robusta]

print(f"  Gạo     — Đỉnh tương quan: r = {best_r_rice:.3f}    tại Lag = {best_lag_rice} tháng")
print(f"  Robusta — Đỉnh tương quan: r = {best_r_robusta:.3f}  tại Lag = {best_lag_robusta} tháng")
```

 ## 2.4 — Trực quan hoá Dữ liệu
 Kết quả định lượng từ bước 2.3 được chuyển hóa thành hai lớp biểu đồ phân tích trực quan:

- **Đường cong tương quan theo độ trễ (Lag Correlation Curve):** Trục hoành là số tháng trễ
  (0 → 12), trục tung là hệ số tương quan Pearson — đỉnh đường cong chỉ ra độ trễ tối ưu.
- **Biểu đồ đa trục (Dual-axis Chart):** Trục trái hiển thị ONI với dải màu cảnh báo El Niño
  cực đoan; trục phải hiển thị đường giá đã dịch pha — cho thấy trực quan sự lệch pha giữa
  hai chuỗi thời gian.

```python
# Lag Correlation Curve
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(lags, corr_rice_lags,    marker='o', color='#e74c3c',
        linewidth=2, label='ONI → Giá Gạo')
ax.plot(lags, corr_robusta_lags, marker='s', color='#2ecc71',
        linewidth=2, label='ONI → Giá Cà phê Robusta')
ax.axvline(x=best_lag_rice,    color='#e74c3c', linestyle='--', alpha=0.6)
ax.axvline(x=best_lag_robusta, color='#2ecc71', linestyle='--', alpha=0.6)
ax.axhline(y=0, color='black', linewidth=0.8)
ax.set_xlabel('Độ trễ (tháng)', fontsize=12)
ax.set_ylabel('Hệ số tương quan Pearson (r)', fontsize=12)
ax.set_title('Đường cong Tương quan theo Độ trễ\nONI vs Giá Gạo & Cà phê Robusta',
             fontsize=13, fontweight='bold')
ax.set_xticks(list(lags))
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('lag_correlation_curve.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
## Dual-axis Chart (ONI & Giá Gạo, dịch pha)
fig, ax1 = plt.subplots(figsize=(14, 6))

# Trục trái: ONI
ax1.set_xlabel('Năm', fontsize=12)
ax1.set_ylabel('Chỉ số ONI', color='#2980b9', fontsize=12)
ax1.plot(df['Date'], df['ONI'], color='#2980b9', linewidth=1.2, alpha=0.8, label='ONI')
ax1.axhline(y=1.5,  color='#e74c3c', linestyle='--', linewidth=1)
ax1.fill_between(df['Date'], df['ONI'], 1.5,
                 where=(df['ONI'] >= 1.5), alpha=0.2, color='#e74c3c',
                 label='El Niño cực đoan (ONI ≥ 1.5)')
ax1.tick_params(axis='y', labelcolor='#2980b9')

# Trục phải: Giá Gạo đã dịch pha
ax2 = ax1.twinx()
ax2.set_ylabel(f'Giá Gạo — USD/tấn (dịch {best_lag_rice} tháng)',
               color='#e67e22', fontsize=12)
ax2.plot(df['Date'], df['Rice'].shift(-best_lag_rice),
         color='#e67e22', linewidth=1.5,
         label=f'Giá Gạo (Lag -{best_lag_rice}M)')
ax2.tick_params(axis='y', labelcolor='#e67e22')

fig.suptitle(f'ONI vs Giá Gạo (dịch pha {best_lag_rice} tháng) | r = {best_r_rice:.3f}',
             fontsize=13, fontweight='bold')
fig.legend(loc='upper left', bbox_to_anchor=(0.08, 0.90), fontsize=9)
plt.tight_layout()
plt.savefig('dual_axis_rice.png', dpi=150, bbox_inches='tight')
plt.show()
```

##3.2 — Xác định đỉnh tương quan và Độ trễ
Thông qua thuật toán dịch chuyển chuỗi thời gian (mục 2.3), mức độ tương quan tăng dần và
đạt **đỉnh cao nhất tại Lag = 6 tháng** với hệ số **r ≈ 0.58** — mức tương quan trung bình
khá có ý nghĩa thống kê trên bộ dữ liệu lịch sử hơn 60 năm.

| Mặt hàng | Độ trễ tối ưu | Hệ số tương quan (r) |
|---|---|---|
| Gạo (Rice) | 6 tháng | ~0.58 |
| Cà phê Robusta | 6 tháng | ~0.52 |

Con số **6 tháng** là định lượng toán học cho khoảng thời gian vật lý cần thiết để một
hiện tượng cực đoan giữa Thái Bình Dương tạo ra cú sốc giá trên thị trường toàn cầu.

##3.3 — Giải thích cơ chế truyền dẫn vĩ mô
Độ trễ 6 tháng tìm được có thể được giải thích qua từng mắt xích của chuỗi cung ứng vật lý:

| Giai đoạn | Thời gian ước tính |
|---|---|
| El Niño bùng phát → Đảo lộn hoàn lưu khí quyển | 1–2 tháng |
| Hạn hán hình thành tại Đông Nam Á / Nam Mỹ | 1–2 tháng |
| Mất mùa, giảm sản lượng thu hoạch | 1–2 tháng |
| Khan hiếm nguồn cung → Giá tăng trên sàn quốc tế | 1–2 tháng |

**Chuỗi nhân quả hoàn chỉnh:**
Nhiệt độ đại dương tăng → Thay đổi hoàn lưu khí quyển → Hạn hán tại vựa nông sản →
